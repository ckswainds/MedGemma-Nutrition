```python
from langchain_ollama import OllamaEmbeddings
embeddings=OllamaEmbeddings(model="llama3")
```

```python
from langchain_chroma import Chroma
vector_store=Chroma(
    collection_name="nutrition_collection",
    embedding_function=embeddings,
    persist_directory="C:\Vscode\git\MedGemma-Nutrition\\assets\\nutrition_vec_db"
)
```

```python
# vector_store.add_documents(docs)
```

```python
def retrieve_context(query:str):
    """Retrieve information to help answer a query."""
    retreived_docs=vector_store.similarity_search(query=query)
    serialized="\n\n".join(
        (f"source:{doc.metadata}\n Content:{doc.page_content}")
        for doc in retreived_docs
    )
    return serialized,retreived_docs
```

```python
from langchain_core.prompts import PromptTemplate
template="""
make a nutrition plan for {category} and here are the information {infos} use only these information to answer,
if you dont know the answer directly say i dont know the answer
"""
prompt=PromptTemplate.from_template(template)
```

```python
query = "diabetes diet plan"
category = "diabetes patient"

# Retrieve context from vector store
serialized_docs, retrieved_docs = retrieve_context(query)

# Format the prompt with retrieved documents
formated_prompt = prompt.format(category=category, infos=serialized_docs)
```

```python
# Stream response token by token
for chunk in llm.stream(formated_prompt):
    print(chunk.content, end="", flush=True)
```
